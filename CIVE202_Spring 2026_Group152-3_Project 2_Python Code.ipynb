```python
import pandas as pd
import numpy as np
```

```python
cubic_yard_ft3 = 27 # cubic feet in one cubic yard
unit_water_weight = 62.4 # lb/ft^3 (unit weight of water)

def water_weight_Q(cement_A,fly_ash_B,silica_fume_C,other_SCM_D,wc_ratio_E):
    total_cementious = cement_A + fly_ash_B + silica_fume_C + other_SCM_D 
    water_weight_Q = total_cementious * wc_ratio_E
    return water_weight_Q

def cement_volume_R(cement_A, gravity_J):
    cement_volume_R = cement_A / (gravity_J * unit_water_weight)
    return cement_volume_R

def fly_ash_volume_S(fly_ash_B, gravity_fly_ash_K):
    fly_ash_volume_S = fly_ash_B / (gravity_fly_ash_K * unit_water_weight)
    return fly_ash_volume_S

def silica_fume_volume_T(silica_fume_C, gravity_silica_fume_L):
    silica_fume_volume_T = silica_fume_C / (gravity_silica_fume_L * unit_water_weight)
    return silica_fume_volume_T

def scm_volume_U(other_scm_D, gravity_scm_M):
    scm_volume_U = other_scm_D / (gravity_scm_M * unit_water_weight)
    return scm_volume_U

def air_volume_V(air_content_F):
    return (air_content_F/100)*cubic_yard_ft3

def water_volume_W(Q):
    return (Q / 62.4)

def total_aggregate_volume_X(R, S, T, U, V, W):
    volume_X = (cubic_yard_ft3 - R - S - T - U - V - W)
    return volume_X

def fine_aggregate_weight_Y(percent_fine_G, sg_fine_N, X):
    weight_Y = unit_water_weight * (percent_fine_G / 100) * sg_fine_N * X
    return weight_Y

def coarse_aggregate_weight_Z(percent_coarse_H, sg_coarse_O, X):
    return unit_water_weight * (percent_coarse_H / 100) * sg_coarse_O * X

def other_aggregate_AA(percent_other_I, sg_other_P, X):
    return unit_water_weight * (percent_other_I / 100) * sg_other_P * X



project_no = int(input("Enter project number: "))
concrete_class = input("Enter class of concrete: ")

# Cementitious material inputs
cement_A = float(input("Enter cement weight A (lb/yd^3):"))
fly_ash_B = float(input("Enter fly ash weight B (lb/yd^3):"))
silica_fume_C = float(input("Enter silica fume weight C (lb/yd^3):"))
other_scm_D = float(input("Enter other SCM weight D (lb/yd^3):"))

# Design parameters
water_cement_ratio_E = float(input("Enter target water-cement ratio E: "))
air_content_F = float(input("Enter target air content F(%): "))

# Aggregate proportions
percent_fine_G = float(input("Enter percent fine aggregate G(%): "))
percent_coarse_H = float(input("Enter percent coarse aggregate H(%): "))
percent_other_I = float(input("Enter percent other aggregate I(%): "))

# Specific gravities
sg_cement_J = float(input("Enter specific gravity of cement J:"))
sg_fly_ash_K = float(input("Enter specific gravity of fly ash K:"))
sg_silica_fume_L = float(input("Enter specific gravity of silica fume L:"))
sg_other_scm_M = float(input("Enter specific gravity of other SCM M:"))

sg_fine_N = float(input("Enter specific gravity of fine aggregate N:"))
sg_coarse_O = float(input("Enter specific gravity of coarse aggregate O:"))
sg_other_P = float(input("Enter specific gravity other aggregate P:"))

# A Print message to confirm all inputs have been collected

print("\nAll user inputs collected successfully.")

Q = water_weight_Q(cement_A, fly_ash_B, silica_fume_C, other_scm_D, water_cement_ratio_E)
R = cement_volume_R(cement_A, sg_cement_J)
S = fly_ash_volume_S(fly_ash_B, sg_fly_ash_K)
T = silica_fume_volume_T(silica_fume_C, sg_silica_fume_L)
U = scm_volume_U(other_scm_D, sg_other_scm_M)
V = air_volume_V(air_content_F)
W = water_volume_W(Q)   
X = total_aggregate_volume_X(R, S, T, U, V, W)
Y = fine_aggregate_weight_Y(percent_fine_G, sg_fine_N, X)
Z = coarse_aggregate_weight_Z(percent_coarse_H, sg_coarse_O, X)
AA = other_aggregate_AA(percent_other_I, sg_other_P, X)

print("\n---------------------------------------------")
print(" NDOT Concrete Mix Design – Weight Summary")
print("         (1 Cubic Yard of Concrete)")
print("---------------------------------------------")
print(f"Project Number:       {project_no}")
print(f"Class of Concrete:    {concrete_class}")
print("---------------------------------------------")
print(f"Cement (A):            {cement_A:8.1f} lb")
print(f"Fly Ash (B):           {fly_ash_B:8.1f} lb")
print(f"Silica Fume (C):       {silica_fume_C:8.1f} lb")
print(f"Other SCM (D):         {other_scm_D:8.1f} lb")
print("---------------------------------------------")
print(f"Fine Aggregate (Y):    {Y:8.0f} lb")
print(f"Coarse Aggregate (Z):  {Z:8.0f} lb")
print(f"Other Aggregate (AA):  {AA:8.0f} lb")
print("---------------------------------------------")
print(f"Water (Q):             {Q:8.0f} lb")
print("---------------------------------------------")
print("End of Mix Design Summary")
```
